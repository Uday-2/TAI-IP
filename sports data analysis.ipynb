```python
#histogram visualization for top4 players total stats
selected_players=['G.Pasquale','Luis García','Iniesta','D.Yorke']
df=pd.DataFrame({'Name':['G.Pasquale','Luis García','Iniesta','D.Yorke'],
                   'Total Stats':[1929,1906,1770,1348]})
#filtering data for selected players
selected_players_data=df[df['Name'].isin(selected_players)]

#ploting histogram
plt.bar(selected_players_data['Name'], selected_players_data['Total Stats'])
plt.xlabel('Name')
plt.ylabel('Total Stats')
plt.title('players performence')
plt.show()
```

```python
#creating a line plot to show performence of multiple players
plt.figure(figsize=(12,5))
players=['G.Pasquale','Luis García','Iniesta','D.Yorke']
agility_scores=[68,62,71,79]
plt.plot(players,agility_scores,marker='o', linestyle='-',color='pink',label='Agility score')
#adding labels and titles
plt.xlabel('players')
plt.ylabel('Agility score')
plt.title('players agility scores')
plt.legend()
plt.show()
```
